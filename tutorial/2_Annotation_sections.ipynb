## Add annotation of the sections in metadata

````python
import pandas as pd
import os
import numpy as np
````

````python
path = os.getcwd()
path=path+"\\cell_ID\\"  # the path where exported cell_ID are saved  
print(path)
````

````python
## A file with section information is provided

info = pd.read_csv(path+"section_ID.csv")
info
````

````python
## merge section infomation with the single cell ID

crop_info = info["crop"]
crop_info=crop_info.tolist()

stage_info = info["stage"]
stage_info=stage_info.tolist()

embryo_ID=info["embryo_ID"].tolist()
section_level=info["level"].tolist()

uniqueID=info["uniqueID"].tolist()

print(crop_info)
print(stage_info)
print(embryo_ID)
print(section_level)  
print(uniqueID)

# get the cell_ID file name

directory_string = os.getcwd()
type( directory_string) 

filename =  [ directory_string + '\\cell_ID\\' + 'crop' + '{:d}'.format(crop_name) + ".csv"  for crop_name in crop_info ]
new_filename = [ directory_string + '\\cell_ID\\' + 'crop' + '{:d}'.format(crop_name) + "_anno.csv" for crop_name in crop_info]

print(filename)

# add section information

for idx_file, file in enumerate( filename ):
    data = pd.read_csv( file )
    #data.rename(columns={ data.columns[0]: "cell_ID" }, inplace = True)
    #add new lines
    data = data.assign(crop=crop_info[ idx_file ] ) 
    data = data.assign(stage=stage_info[ idx_file ])
    data = data.assign(embryo_ID=embryo_ID[ idx_file ])
    data = data.assign(section_level=section_level[ idx_file ])
    data = data.assign(uniqueID=uniqueID[ idx_file ])
    #save new files
    data.to_csv( new_filename[ idx_file ], index=False)
    
for idx_newfile, newfile in enumerate( new_filename ):
    data = pd.read_csv( newfile )
    data.rename(columns={ data.columns[0]: "EntityID" }, inplace = True)
    if(idx_newfile == 0):
        data_all = data.copy()
    else:
        data_all = pd.concat((data_all, data),axis=0 )
                                  
        
data_all_dataframe = pd.DataFrame(data_all)
data_all.to_csv("EntityID_anno.csv", index=False)
````

````python
## add cell_metadata info based on cell ID


EntityID = pd.read_csv("EntityID_anno.csv") 
metadata=pd.read_csv("cell_metadata.csv")

metadata_anno = pd.merge(EntityID,metadata,on='EntityID', how='left')  ## only keep the metadata of selected cells
metadata_anno.to_csv("cell_metadata_anno.csv",index = False)

print(f"metadata:{metadata.shape}") 
print(f"metadata_anno:{metadata_anno.shape}")

````

````python
## filter cell_by_gene matrix based on metadata_anno
## get the gene expression matrix with the same EntityID of metadata_anno

matrix=pd.read_csv("cell_by_gene.csv")


matrix_filtered = matrix[matrix['cell'].isin(metadata_anno['EntityID'])]
matrix_filtered
matrix_filtered.to_csv("cell_by_gene_filtered.csv", index=False)


print(f"matrix:{matrix.shape}")
print(f"matrix_filtered:{matrix_filtered.shape}")   ## matrix_filtered should be the same shape as metadata_anno
````
